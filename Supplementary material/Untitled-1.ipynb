```python
import pandas as pd
```

```python
bench_df=pd.read_table("benchmarking_set_parsed_totc.tsv", names=["SP", "Seq", "Class","AC"])
bench_df.drop(columns="SP")
bench_df=bench_df[["AC", "Seq", "Class"]]
bench_df["CV"]="Nan"
bench_df["Benchmarking"]=1
bench_df
```

```python
df_cv1=pd.read_table("SMCV1", names=["SP", "Seq", "Class","AC"])
df_cv1.drop(columns="SP")
df_cv1=df_cv1[["AC", "Seq", "Class"]]
df_cv1["CV"]=0
df_cv1["Benchmarking"]=0

df_cv2=pd.read_table("SMCV2", names=["SP", "Seq", "Class","AC"])
df_cv2.drop(columns="SP")
df_cv2=df_cv2[["AC", "Seq", "Class"]]
df_cv2["CV"]=1
df_cv2["Benchmarking"]=0

df_cv3=pd.read_table("SMCV3", names=["SP", "Seq", "Class","AC"])
df_cv3.drop(columns="SP")
df_cv3=df_cv3[["AC", "Seq", "Class"]]
df_cv3["CV"]=2
df_cv3["Benchmarking"]=0

df_cv4=pd.read_table("SMCV4", names=["SP", "Seq", "Class","AC"])
df_cv4.drop(columns="SP")
df_cv4=df_cv4[["AC", "Seq", "Class"]]
df_cv4["CV"]=3
df_cv4["Benchmarking"]=0

df_cv5=pd.read_table("SMCV5", names=["SP", "Seq", "Class","AC"])
df_cv5.drop(columns="SP")
df_cv5=df_cv5[["AC", "Seq", "Class"]]
df_cv5["CV"]=4
df_cv5["Benchmarking"]=0

df_cv1
```

```python
entry_table=pd.concat([df_cv1,df_cv2,df_cv3,df_cv4,df_cv5,bench_df], axis=0)
```

```python
entry_table= entry_table.reset_index(drop=True)
entry_table= entry_table.rename(columns={"AC":"Accession code", "Seq":"Sequence","Class":"Signal peptide","CV":"CV", "Benchmarking":"Benchmarking"})
entry_table
entry_table.to_csv("entries_table.tsv",sep="\t")
```
